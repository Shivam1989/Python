# Calculate the frequency for each words and show the top 10 most frequent words and their frequency.


```python
#Getting all the word from the file
file=nltk.corpus.gutenberg.words('austen-persuasion.txt')
# I created a Dictionary to store all unique words.
wordcount={}
#Loop to find the unique words
for word in file:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
print (word,wordcount)


```

```python
#Length of unique Words 
uniqueWords = len(wordcount)
print('Count of Unique words', uniqueWords)
```

# Sorted Frequency of each words

```python
# sorted is used to sort dictionary in descending order.
sortedFrequency = sorted(wordcount.items(), key=lambda x: (-x[1], x[0])) 
print(sortedFrequency)
```
