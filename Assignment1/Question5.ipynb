# Use Gutenberg and Web_text data. Find out what are the top 5 words that Shakespeare used but we do not use in currently

```python
from nltk.corpus import gutenberg, webtext
from nltk.corpus import stopwords
```

```python
webtext.fileids()
```

```python
#Finding all the field ids from gutenberg
F1=gutenberg.fileids()
#find all words in gutenberg to find top 50 words 
F2=list(gutenberg.words())
#Retrieving all the files with name starting with shakespeare
shake = [elem for elem in F1 if elem.startswith('shakespeare')]
print(shake)
```

```python
# Defining function to remove punctuations
import re
def punctuation(elem):
    pattern=re.compile('\W+')
    return  re.match(pattern,elem)

webtext_corpus=list(webtext.words())

newList1=[elem1 for elem1 in webtext_corpus if not punctuation(elem1) ]

#Converting all elements to lower case
all_posts = ([word1.lower() for word1 in newList1])

#Removing the Stopwords from the list

filtered_all_posts = all_posts[:] #make a copy of the word_list
for word in all_posts: # iterate over word_list
    if word in stopwords.words('english'): 
        filtered_all_posts.remove(word) # remove word from filtered_word_list if it is a stopwor
```

```python
#Creating list to get all the words from shakespeare txt file
shakeWords = []
for file in shake:
    shakeWords.extend(gutenberg.words(file))
    
newList2=[elem1 for elem1 in shakeWords if not punctuation(elem1) ]
    
#Converting all elements to lower case
shakeWords= ([word.lower() for word in newList2])




#Removing the Stopwords from the list

filtered_shakeWords = shakeWords[:] #make a copy of the word_list
for word in all_posts: # iterate over word_list
    if word in stopwords.words('english'): 
        filtered_shakeWords.remove(word) # remove word from filtered_word_list if it is a stopwor
```

```python
result=([word for word in filtered_shakeWords if word not in filtered_all_posts])
```

```python
wordcount={}
#Loop to find the unique words
for word in result:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
#print (word,wordcount)

#sorted is used to sort dictionary in descending order.
sortedFrequency = sorted(wordcount.items(), key=lambda x: (-x[1], x[0])) 
print(sortedFrequency)
```

# Top 5 Words used shakespeare but not frequent

```python
# Showing the Top 5 Frequent words used in file
sortedFrequency[0:5]
```
