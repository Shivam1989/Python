```python
#Importing Corpus
from nltk.corpus import nps_chat, words
```

```python
#Finding Wordlist
word_corpus=list(words.words())
word_corpus = set([word.lower() for word in word_corpus])
```

```python
#Finding Uncommon Words
all_posts = list(nps_chat.words())
all_posts = set([word1.lower() for word1 in all_posts])
```

```python
#Obtaining result
result=([word for word in all_posts if word not in word_corpus ])
```

```python
print(result)
```
